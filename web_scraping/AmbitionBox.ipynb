```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage = requests.get('https://www.ambitionbox.com/jobs?campaign=desktop_nav',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
company = soup.find_all('div',class_='jobsInfoCardCont')
```

```python
len(company)
```

```python
name = []
rating = []
rev = []

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_='rating_text rating_text--sm').text.strip())
    text = i.find('a', class_='review-count caption-strong-medium').text.strip()
    rev.append(text.split()[0].replace('(', ''))
    

d={'name':name,'rating':rating,'rev':rev}
df = pd.DataFrame(d)   
df
```
