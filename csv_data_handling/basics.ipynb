<span style="color:orange"><b>Convertors</b></span>

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    if name == "Mumbai Indians":
        return "MI"
    else:
        return name 
```

```python
rename("Royal Challengers Bangalore")
rename("Mumbai Indians")
```

```python
df = pd.read_csv(url,converters={'team1':rename,'team2':rename})
```

```python
df
```
